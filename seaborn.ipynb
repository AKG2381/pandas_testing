```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the tips dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python


# 1. Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='day', y='tip', data=tips, estimator='mean', palette='Blues')
plt.title('Average Tip by Day')
plt.show()
```

```python
# 2. Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='Set2')
plt.title('Total Bill Distribution by Day and Smoker Status')
plt.show()

```

```python
# 3. Heatmap
numeric_cols = tips.select_dtypes(include=['float64', 'int64'])
# Plot the heatmap with only numeric columns
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Columns in Tips Dataset')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical columns to numeric values
label_encoder = LabelEncoder()
categorical_cols = ['sex', 'smoker', 'day', 'time']
for col in categorical_cols:
    tips[col] = label_encoder.fit_transform(tips[col])

# Plot the heatmap with encoded categorical columns
plt.figure(figsize=(8, 6))
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Encoded Tips Dataset')
plt.show()
```
